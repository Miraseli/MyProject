```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
fig, ax = plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(20,10))
ax[0,0].hist(data_1['age'], bins=25, facecolor='green', edgecolor='black')
ax[0,0].set(title='age');
ax[0,1].hist(data_1['items'], bins=15)
ax[0,1].set(title='items');
ax[1,0].hist(data_1['amount'], bins=30, facecolor='orange', edgecolor='black')
ax[1,0].set(title='amount');
ax[1,1].hist(data_1['region'], facecolor='red', edgecolor='black')
ax[1,1].set(title='region');
fig.suptitle('EXAMINATION OF EACH ELEMENT WITH HISTOGRAM', fontsize=20, fontweight='bold');
```

```python
import seaborn as sns
```

```python
# Using seaborn to plot the count of purchase in each region
fg = sns.catplot(x='region', data=data_1, kind='count', aspect=2)
fg.set_xticklabels(["North", "South", "East","West"])
fg.set_xlabels('region')
```

As shown in the figure above West region made the most number of purchase

### Amount spend in each region

```python
# how much costumer spent in each region
data_region_amount = data_1.iloc[:,3:]
data_region_amount
```

```python
data_region_amount.groupby('region').describe()
```

```python
data_region_amount.groupby('region').sum()
```

```python
# Using seaborn to plot the amount spend in each region
fg_1 = sns.catplot(x='region', y='amount', data=data_region_amount, kind="bar")
fg_1.set_xticklabels(["North", "South", "East","West"])
fg_1.set_xlabels('region')
fg_1.set_xlabels('amount')
```

As shown in the figure above West region spent the greatest amount of money, and Sount the smallest amount of money.
